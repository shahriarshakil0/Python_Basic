```python
def student_info(*args, **kwargs):
  print(args)
  print(kwargs)

student_info('bangla','english','ict', name=['Shahriar','shakil'], age=[22,24])
```

```python
def leap_year(year):
  if int(year) %2 == 0:
    print('Leap Year!')
  else:
    print('Not a leap year!')

leap_year(2021)
```
