```python
if True:
  print('Shakil')
```

```python
name = 'shakil'
if name == 'shakil':
  print('that is my name')
else:
  print('not my name')
```

```python
name = 'messi'
if name == 'shakil':
  print('that is my name')
elif name == 'messi':
  print('this is messi')
else:
  print('not my name')
```

```python
user = 'Admin'
login = True

if user == 'Admin' and login:
  print('successfully login')
else:
  print('not authorised')
```

```python
user = 'Admin'
login = False

if user == 'Admin' and login:
  print('successfully login')
else:
  print('not authorised')
```

```python
user = 'Admin'
login = False

if user == 'Admin' and not login:
  print('please login')
else:
  print('welcome')
```
